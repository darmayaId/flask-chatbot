```python
import json
import pandas as pd

# Memuat data dari file JSON dengan encoding UTF-8
with open('intents.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# Konversi data menjadi DataFrame
df = pd.DataFrame(data)

# Print the DataFrame to verify the result
df.head(50)
```

Prepocessing

1. Case Folding

Proses mengubah huruf pada kalimat menjadi huruf kecil

```python
# Langkah 1: Case folding

# Ekstrak data dari JSON ke dalam DataFrame
intents_list = []

for intent in data['intents']:
    for pattern in intent['patterns']:
        intents_list.append({'pattern': pattern, 'tag': intent['tag']})

df = pd.DataFrame(intents_list)

# Langkah 1: Case folding
df['pattern'] = df['pattern'].str.lower()

df.head(50)
```

2. Tokenizing

Proses pemisahan teks menjadi potongan kata

```python
df['pattern'] = df['pattern'].str.lower()

# Tokenizing kolom 'pattern' menggunakan word_tokenize
df['tokenized_pattern'] = df['pattern'].apply(word_tokenize)

# Menampilkan hasil 50 data teratas setelah tokenizing
df[['pattern', 'tokenized_pattern']].head(50).to_string(index=False)
df.head(50)
```

3. Stopword Removal

Penghapusan kata yang sering muncul namun diabaikan

```python
# Langkah 3: Stopword removal
stopwords = ['harus', 'seperti', 'untuk', 'siapa', 'mengapa', 'dimana', 'apakah', 'saja', 'itu', 
             'yang', 'dan', 'di', 'juga', 'adalah', 'dengan', 'sebagai', 'berapa', 'apa', 'ini', 
             'an', 'ada', 'dari', 'alasan']

def remove_stopwords(tokens):
    return [word for word in tokens if word.lower() not in stopwords]

# Terapkan fungsi remove_stopwords pada kolom 'tokenized_pattern'
df['cleaned_pattern'] = df['tokenized_pattern'].apply(remove_stopwords)

# Menampilkan hasil 50 data teratas setelah stopword removal
df[['pattern', 'tokenized_pattern', 'cleaned_pattern']].head(50).to_string(index=False)
df.head(50)
```

4. Proses Stemming

Proses pengubahan kata kerja menjadi kata dasar

```python
# Langkah 4: Stemming

factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
# Fungsi untuk melakukan stemming
def stem_words(tokens):
    return [stemmer.stem(word) for word in tokens]

# Terapkan fungsi stemming pada kolom 'cleaned_pattern'
df['stemmed_pattern'] = df['cleaned_pattern'].apply(stem_words)

# Menampilkan hasil 50 data teratas setelah stemming
df[['pattern', 'cleaned_pattern', 'stemmed_pattern']].head(50).to_string(index=False)
df.head(50)
```

Tokenisasi

Tujuan: Untuk menyimpan tokenizer sehingga dapat digunakan kembali nanti tanpa perlu memproses ulang teks.
Ini adalah langkah awal yang umum dalam pipeline pemrosesan teks sebelum melatih model machine learning atau deep learning pada data teks.

```python
from tensorflow.keras.preprocessing.text import Tokenizer
import pickle
import pandas as pd

#apply(lambda x: ' '.join(x)) akan menggabungkan kata-kata dalam setiap 
#entri dalam kolom stemmed_pattern menjadi sebuah kalimat utuh (string).
texts = df['stemmed_pattern'].apply(lambda x: ' '.join(x)).tolist()

# Inisialisasi dan fit tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

# Konversi teks menjadi urutan token
sequences = tokenizer.texts_to_sequences(texts)

# Ambil daftar kata unik
word_index = tokenizer.word_index
unique_words = list(word_index.keys())

# Asumsikan kolom 'tag' sebagai label untuk setiap pattern
labels = df['tag'].tolist()

# Menyimpan tokenizer dan data yang diperlukan
with open('tokenizer.pkl', 'wb') as handle:
    pickle.dump(tokenizer, handle, protocol=pickle.HIGHEST_PROTOCOL)

print("Tokenizer and sequences created successfully!")

```

```python
# Menggabungkan pola dan intent ke dalam dokumen
documents = list(zip(df['stemmed_pattern'], df['tag']))
print(f"{len(documents)} documents")

# Menentukan intent sebagai kelas
classes = sorted(set(df['tag']))
print(f"{len(classes)} classes: {classes}")

# Mengambil semua kata unik yang telah distem
words = sorted(set([word for sublist in df['stemmed_pattern'] for word in sublist]))
print(f"{len(words)} unique lemmatized words: {words}")

# Menyimpan Python objects ke file pickle yang akan digunakan saat prediksi
pickle.dump(words, open('tek.pkl', 'wb')) 
pickle.dump(classes, open('lab.pkl', 'wb'))
```

Bag of Words


Proses menghitung kemunculan setiap kata dengan melakukan training

```python
import numpy as np
import random
from nltk.stem import WordNetLemmatizer

# Inisialisasi lemmatizer
lemmatizer = WordNetLemmatizer()

# Buat data pelatihan
training = []

# Buat array kosong untuk output
output_empty = [0] * len(classes)

# Proses set pelatihan, bag of words untuk setiap kalimat
for doc in documents:
    # Inisialisasi bag of words
    bag = []
    # Daftar kata tokenisasi untuk pola
    pattern_words = doc[0]
    
    # Lemmatize setiap kata - buat kata dasar
    pattern_words = [lemmatizer.lemmatize(word.lower()) for word in pattern_words]
    
    # Buat array bag of words dengan 1 jika kata cocok ditemukan dalam pola saat ini
    for w in words:
        bag.append(1) if w in pattern_words else bag.append(0)
    
    # Output adalah '0' untuk setiap tag dan '1' untuk tag saat ini (untuk setiap pola)
    output_row = list(output_empty)
    output_row[classes.index(doc[1])] = 1
    training.append([bag, output_row])

# Acak fitur dan konversi ke array NumPy
random.shuffle(training)
training_array = np.array(training, dtype=object)
training = training_array


# Menampilkan beberapa hasil pelatihan
print("Contoh Data Pelatihan:")
for i in range(min(len(training), 5)):  # Tampilkan hingga 5 contoh
    print(f"Input (Bag of Words): {training[i][0]}")
    print(f"Output (Label): {training[i][1]}")
    print()

```

```python
# Buat daftar pelatihan dan pengujian
train_x = list(training[:,0])
train_y = list(training[:,1])
# Menampilkan beberapa hasil pelatihan
print("Contoh Data Pelatihan:")
for i in range(min(len(train_x), 5)):  # Tampilkan hingga 5 contoh
    print(f"Input: {train_x[i]}")
    print(f"Output: {train_y[i]}")
    print()
```

Pemodelan machine learning 

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import SGD

# Set seed untuk konsistensi hasil
np.random.seed(42)
tf.random.set_seed(42)
random.seed(42)
#Dalam konteks ini, angka 42 digunakan secara humoris sebagai nilai default atau placeholder.
# Buat model Neural Network
model = Sequential()
model.add(Dense(128, input_shape=(len(train_x[0]),), activation='relu'))
#ReLU (Rectified Linear Unit) adalah fungsi aktivasi yang umum digunakan untuk layer tersembunyi karena kemampuannya untuk  mencegah masalah vanishing gradient.
model.add(Dropout(0.5))
#Dropout adalah teknik regularisasi yang membantu mencegah overfitting
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(train_y[0]), activation='softmax'))
#Softmax digunakan untuk klasifikasi multi-kelas. Ini menghasilkan probabilitas untuk setiap kelas

# Kompilasi model
#algoritma optimisasi untuk melatih model Neural Network. Ini adalah metode yang mengupdate bobot model berdasarkan gradien dari fungsi loss.
sgd = SGD(learning_rate=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])

# Latih model
hist = model.fit(np.array(train_x), np.array(train_y), epochs=150, batch_size=5, verbose=1)
#Jumlah Epoch: 150 epoch adalah pilihan yang cukup umum, namun jumlah ini bisa bervariasi berdasarkan dataset dan kompleksitas model. Nilai ini dipilih untuk memberikan model cukup waktu untuk belajar dari data.
#Batch Size: Batch size 5 berarti setiap update bobot dilakukan berdasarkan 5 sampel data sekaligus. Ukuran batch mempengaruhi performa pelatihan dan kecepatan konvergensi.
#verbose=1: Menampilkan progress bar dan informasi tentang epoch, termasuk loss dan metrik evaluasi seperti akurasi. Ini berguna untuk memantau pelatihan dan mengidentifikasi masalah jika ada.

# Simpan model
model.save('modelas.h5')

print("\n" + "*" * 50)
print("\nModel Created Successfully!")


```
